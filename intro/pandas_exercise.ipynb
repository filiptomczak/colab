```python
df=fetch('MSFT')

df.info()
```

```python
df.head()
```

```python
avg = (df['open']+df['close'])/2
df['avg']=avg

df.columns=[col.lower() for col in df.columns]
```

```python
df.head()
```

```python
df=df.sort_index()
df.tail()
```

```python
df['close_shift']=df[['close']].shift(1)
df['daily_change']=df['close']/df['close_shift']-1
```

```python
df.head()
```

```python
df.daily_change.hist(bins=100)
```

```python
df.close.plot()
```

```python
df_pos= df[df.daily_change>0]
df_neg= df[df.daily_change<0]

print(df_pos['daily_change'].mean())
print(df_neg['daily_change'].mean())
```
