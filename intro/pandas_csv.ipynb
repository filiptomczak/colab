```python
def fetch(company='amzn'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

apple=fetch('aapl')
amazon=fetch()
uber=fetch('uber')
meta=fetch('meta')
```

```python
apple.info()
```

```python
uber.info()
```

```python
apple.columns = ['apple_'+col.lower() for col in apple.columns]
uber.columns = ['uber_'+col.lower() for col in uber.columns]
amazon.columns = ['amazon_'+col.lower() for col in amazon.columns]
meta.columns = ['meta_'+col.lower() for col in meta.columns]
```

```python
meta.head()
```

```python
df=pd.concat(objs=[apple,amazon,meta,uber],axis=1)
```

```python
pd.set_option('display.float_format',lambda x:f'{x:.2f}')
df.describe().T
```

```python
closes=[col for col in df.columns if col.endswith('close')]
closes
```

```python
df[closes].corr()
```

```python
opens=[col for col in df.columns if col.endswith('open')]

co=df[closes].append(df[opens])
co
```
