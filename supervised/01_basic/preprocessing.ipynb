```python
import sklearn
import numpy as np
import pandas as pd
```

```python
data={
    'size':['xl','l','m','l','m'],
    'color':['red','green','blue','green','red'],
    'gender':['female','male','male','female','female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw=pd.DataFrame(data=data)
df_raw
```

```python
df=df_raw.copy()
df.info()
```

```python
for col in ['size','color','gender','bought']:
  df[col]=df[col].astype('category')

df['weight']=df['weight'].astype('float')
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

mapowanie wartosci [yes,no] kolumny bought na [1,0]

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
df['bought'] = le.fit_transform(df['bought'])

df
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder=OneHotEncoder(drop='first',sparse_output=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
pd.get_dummies(data=df)
```

```python
pd.get_dummies(data=df,drop_first=True)
```

```python
pd.get_dummies(data=df,drop_first=True,prefix='new',prefix_sep='-')
```

```python
pd.get_dummies(data=df,drop_first=True,prefix='new',prefix_sep='-',columns=['size'])
```

```python
def standardize(x):
  return (x-x.mean())/x.std()

standardize(df['price'])
```
