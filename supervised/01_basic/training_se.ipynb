```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns

np.random.seed(42)
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x:f'{x:.2f}'))
sns.set()
```

```python
from sklearn.datasets import load_iris
raw_data=load_iris()
raw_data_copy=raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

```python
data=raw_data_copy['data']
target=raw_data_copy['target']

print(f'{data[:5]}\n')
print(f'{target[:5]}')
```

```python
all_data=np.c_[data,target]
all_data[:5]
```

```python
df=pd.DataFrame(data=all_data, columns=raw_data.feature_names+['target'])
df.head()
```

```python
df.info()
```

```python
df.describe().T.apply(lambda x:round(x,2))
```

```python
df.target.value_counts()
```

```python
data=df.copy()
target=data.pop('target')
data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data,target,random_state=42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
from sklearn.datasets import load_breast_cancer

raw_data=load_breast_cancer()
raw_data_copy=raw_data.copy()
raw_data_copy.keys()
```

```python
data=raw_data_copy['data']
target=raw_data_copy['target']

all_data=np.c_[data,target]
all_data[:5]
```

```python
df=pd.DataFrame(data=all_data,columns=list(raw_data['feature_names'])+['target'])
df.head()
```

```python
df.target.value_counts()
```

```python
data=df.copy()
target=data.pop('target')
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(data,target, test_size=0.2,random_state=42,stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```
